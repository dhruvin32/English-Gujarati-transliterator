Function to convert english text into gujarati

```python
def getgujarati(text):

    word = text.strip()
    curr = ''
    token = ''
    l = len(word)
    index = 0
    while index < l:
        
        #decide token size for consonants
        
        if l - index >= 3:
            tokensize = 3
        else:
            tokensize = l - index
        
        isvowel = 0 #to check character is vowel or consonant
        
        #finding mapping of corresponding dictionary
        while(tokensize > 0):
            token = word[index:index+tokensize]
            if token in guj_consonants:
                curr = curr + guj_consonants.get(token)
                index = index + tokensize
                break
            elif token in guj_vowel:
                curr = curr + guj_vowel.get(token)
                index = index + tokensize
                isvowel = 1
                break
            else:
                tokensize-=1
                
        #if its vowel then it wouldn't be followed by any vowel sign
        if isvowel == 1:
            continue
        
        if(index < l):
            nextchar = word[index]
            #print(nextchar)
            
        if l - index >= 2:
            tokensize = 2
        else:
            tokensize = l - index
            
        presentvowel = 0
        
        #loop to map vowel sign
        while(tokensize > 0):
            token = word[index:index+tokensize]
            if token == 'a':
                presentvowel = 1
                index += 1
            else:
                if token in guj_vowelsign:
                    curr = curr + guj_vowelsign.get(token)
                    index = index + tokensize
                    presentvowel = 1
                    break
                else:
                    tokensize-=1
        
        #add nukta if its half character
        if presentvowel == 0 and index < l:
            curr = curr + '\u0ACD' #nukta
            
    return curr
```

Function to convert gujarati text into latin

```python

#defines range for every gujarati symbols

consonants = '\u0A95-\u0AB9\u0AD0\u0AE0'
vowels = '\u0A85-\u0A94'
vowelsigns = '\u0ABE-\u0ACC\u0A81-\u0A83'
nukta = '\u0ACD'
gujaratichars = '\u0A80-\u0AFF'


def getlatin(text):

    word = text.strip()
    curr = ''

    for index, char in enumerate(word): 
        
        if re.match('[' + gujaratichars + ']', char):
            #print(1,end=" ")
            if re.match('[' + consonants + ']', char):
                #print(2,end=" ")
                nextchar = word[(index + 1) % len(word)]
                
                if nextchar:
                    #print(3,end=" ")
                    if re.match('[' + nukta + ']', nextchar):
                        trans = conversiontable.get(char, '')
                        curr = curr + trans
                        continue
                        """print(4,end=" ")
                        cons = char + nextchar
                        nukta_present = 1"""
                    else:
                        #print(5,end=" ")
                        cons = char
                        nukta_present = 0

                    if nukta_present:
                        #print(6,end=" ")
                        nextchar = word[(index + 2) % len(word)]

                    if re.match('[' + vowelsigns +']', nextchar) or index == len(word)-1:
                        #print(7,end=" ")
                        trans = conversiontable.get(cons, '')
                        curr = curr + trans
                    else:
                        #print(8,end=" ")
                        trans = conversiontable.get(cons, '')
                        if not re.match('[' + vowels + ']', cons):
                            trans = trans + "a"
                        curr = curr + trans
            else:
                #print(9,end=" ")
                trans = conversiontable.get(char, '')
                curr = curr + trans

        else:
            #print(10,end=" ")
            curr = curr + char
        #print()

    return curr
```

```python
englisharray = ["ohm","namo","bhagavate","vasudevay","namah","mayur","daanav","utsav","vatsal","udayaaom","dhruvin","zyaadaa"]
gujaratiarray = []
latinarray = []
for i in englisharray:
    gujaratiarray.append(getgujarati(i))

for i in gujaratiarray:
    latinarray.append(getlatin(i))
```
